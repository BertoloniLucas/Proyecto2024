```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%pip install seaborn
import seaborn as sns
```

```python
data = pd.read_csv('./StudentsPerformance 2.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
#transformo en numerico las columnas que son categoricas, en 0 y 1
data["gender"] = le.fit_transform(data["gender"])
data["lunch"] = le.fit_transform(data["lunch"])
data["test preparation course"] = le.fit_transform(data["test preparation course"])
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

data = pd.concat([data, pd.get_dummies(data["race/ethnicity"]).apply(lambda x: x.astype(int))], axis=1)

```

```python
data = data.drop('race/ethnicity', axis=1)
data
```

```python
data["parental level of education"].unique()
```

```python
data.columns
```

```python
# Defino el mapeo de los niveles de educación a valores numéricos
education_mapping = {
    "some high school": 0,
    "high school": 1,
    "some college": 2,
    "associate's degree": 3,
    "bachelor's degree": 4,
    "master's degree": 5
}
# Aplico el mapeo 
data['parental level of education'] = data['parental level of education'].map(education_mapping)
data
```

```python
y = data ['math score']
y
```

```python
# excluimos la columna math score para obtener las x de nuestra función
X = data.drop('math score', axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

```

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)

y_lr_train_pred
```

```python
y_lr_test_pred
```

```python
#rendimiento del modelo
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score (y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

```

```python
print(f"Train MSE: {lr_train_mse}")
print(f"Test MSE: {lr_train_r2}")
print(f"Train R2: {lr_test_mse}")
print(f"Test R2: {lr_test_r2}")
```
