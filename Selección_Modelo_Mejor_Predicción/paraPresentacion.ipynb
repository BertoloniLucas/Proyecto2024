```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV   
from sklearn.metrics import roc_auc_score, roc_curve, accuracy_score
import numpy as np
```

```python
df = pd.read_csv('../data.csv')
```

```python
# Valores ROC AUC de los modelos
roc_auc_train_model_rf = 0.893
roc_auc_test_model_rf = 0.788
roc_auc_train_model_xg = 0.936
roc_auc_test_model_xg = 0.873

# Nombres de los modelos
labels = ["Random forest", "XGBoost"]

# Valores para el entrenamiento y la prueba
train_values = [roc_auc_train_model_rf, roc_auc_train_model_xg]
test_values = [roc_auc_test_model_rf, roc_auc_test_model_xg]
```

```python
# Posiciones de las barras
x = np.arange(len(labels))  # Posición para cada modelo
width = 0.35  # Ancho de las barras

# Crear el gráfico
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, train_values, width, label='Train', color='blue')
bars2 = ax.bar(x + width/2, test_values, width, label='Test', color='orange')

# Añadir etiquetas y título
ax.set_ylabel('ROC AUC')
ax.set_title('Comparación de ROC AUC entre Modelos')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Mostrar el gráfico
plt.show()
```
